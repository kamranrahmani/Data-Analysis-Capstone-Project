## Demo: How to work with database


Download database file.


Connect to the database.


```python
import sqlite3
import seaborn as sns
import matplotlib.pyplot as plt
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
all_master_rows = 'SELECT * FROM master'
df = pd.read_sql_query(all_master_rows, conn)
```

```python
import numpy as np
bins = np.arange(0, 2000000, 250000)
plt.hist(df['ConvertedComp'], bins=bins, edgecolor='black')
plt.xlabel('ConvertedComp')
plt.ylabel('Frequency')
plt.title('Histogram of ConvertedComp')
plt.grid(True)
plt.ticklabel_format(style='plain', axis='x')
plt.xticks(rotation=45)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
plt.figure(figsize=(8, 6))

sns.boxplot(data=df['Age'])
plt.title('Boxplot of Age')
plt.xlabel('Age')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='WorkWeekHrs', data=df)
plt.title('Scatter Plot of Age vs WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.show()

df[['WorkWeekHrs','Age']].corr()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
plt.figure(figsize=(8, 6))
sns.scatterplot(x='WorkWeekHrs', y='CodeRevHrs', size='Age', data=df, sizes=(20, 200), legend=False)
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs')
plt.xlabel('Work week Hrs')
plt.ylabel('Code Rev Hrs')

plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
database_desire_table = 'SELECT DatabaseDesireNextYear, COUNT(DatabaseDesireNextYear) as Number_of_Respondents FROM DatabaseDesireNextYear GROUP BY DatabaseDesireNextYear ORDER BY COUNT(DatabaseDesireNextYear) DESC'
df_db = pd.read_sql_query(database_desire_table,conn)
top5 = df_db.head()
plt.figure(figsize=(8, 6))  # Set the figure size (width, height)

plt.pie(top5['Number_of_Respondents'], labels=top5['DatabaseDesireNextYear'], autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of top 5 database for Next Year')

plt.show()

```

```python
dbs = 'SELECT * FROM DatabaseWorkedWith'
db_df = pd.read_sql_query(dbs, conn)
grouped =db_df.groupby('Respondent')['DatabaseWorkedWith'].unique().reset_index()

only_mysql_users = grouped[grouped['DatabaseWorkedWith'].apply(lambda x: len(x) == 1 and x[0] == 'MySQL')]

print(only_mysql_users.shape)
```

```python
dev_type = 'SELECT DevType, COUNT(*) FROM DevType GROUP BY DevType ORDER BY COUNT(DevType) DESC'
types = pd.read_sql_query(dev_type, conn)
types.head()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
filtered_df = df[(df['Age'] >= 30) & (df['Age'] <= 35)][['Age', 'WorkWeekHrs', 'CodeRevHrs']]
medians = filtered_df.groupby('Age').median()

plt.figure(figsize=(10, 6))
plt.bar(medians.index, medians['WorkWeekHrs'], label='WorkWeekHrs', color='blue')
plt.bar(medians.index, medians['CodeRevHrs'], bottom=medians['WorkWeekHrs'], label='CodeRevHrs', color='orange')

plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.legend()

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
filtered_df1 = df[(df['Age'] >= 25) & (df['Age'] <= 30)][['Age', 'ConvertedComp']]
medians1 = filtered_df1.groupby('Age').median().reset_index()
plt.figure(figsize=(8, 6))
plt.plot(medians1['Age'], medians1['ConvertedComp'], marker='o', linestyle='-', color='b', label='median of Converted Comp')
plt.title('Line Chart Example') 
plt.xlabel('Age')  
plt.ylabel('Median of ConvertedComp')  
plt.legend()

plt.grid(True)  
plt.tight_layout()

plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
main_branch = df['MainBranch'].to_frame().value_counts().to_frame().reset_index()
plt.figure(figsize=(8, 6))  
plt.bar(main_branch['MainBranch'], main_branch['count'], color='skyblue')  
plt.title('Main Branch bar chart') 
plt.xlabel('Categories')  
plt.ylabel('Values')  
plt.grid(True)  
plt.xticks(rotation=45) 
plt.tight_layout()  

plt.show()  
```
